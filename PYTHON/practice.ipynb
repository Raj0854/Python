```python
#Leap Year Checker: Task: This program prompts the user to enter a year and checks if it's a leap year or not. Sample Output:

try:
    year= int(input("Enter a year to check."))
    print("Entered year is : ",year)
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print(f"Result : {year} is a leap year.")
    else:
        print(f"Result : {year} is not a leap year.")   

except:
    print("Error! : Please enter a valid year.")
```
