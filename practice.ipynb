```python
# Handling mixed data types in a list
mixed_list = [1, "2", 3.0, "4.5"]

converted_list = [float(item) if isinstance(item, str) else item for item in mixed_list]
print(converted_list)  # Output: [1, 2.0, 3.0, 4.5]
```

```python
#Odd or Even Checker: Task: This program prompts the user to enter a number and determines whether it's odd or even. Sample Output:
try:
    number = int(input("Enter number to check..."))
    print("Entered number is : ",number)
    if number %2==0:
        print(f"Result : {number} is even number.")
    else:
        print(f"Result : {number} is odd number.")
except:
    print("Error! : Please enter a valid number.")
```

```python
#Age Classifier: Task: This program prompts the user to enter their age and classifies them into different age groups. Sample Output:

try:
    Age = int(input("Enter the age to know your age group."))
    print("Entered age is : ",Age)
    if Age<=13:
        print(f"Result : You are a kid.")
    elif Age<=19:
        print(f"Result : You are a teenager.")
    elif Age<=39:
        print(f"Result : You are a adult.")
    else:
        print(f"Result : You are a senior.")
except:
    print("Error! : Please enter a valid age.")
```

```python
#Leap Year Checker: Task: This program prompts the user to enter a year and checks if it's a leap year or not. Sample Output:

try:
    year= int(input("Enter a year to check."))
    print("Entered year is : ",year)
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print(f"Result : {year} is a leap year.")
    else:
        print(f"Result : {year} is not a leap year.")   

except:
    print("Error! : Please enter a valid year.")
```
