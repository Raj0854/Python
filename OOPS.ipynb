```python
class Account:
    def __init__(self,balance,account_no):
        self.balance=balance
        self.account_no=account_no
    def credit(self,amt):
        self.balance+=amt
        print("your account number is : ",self.account_no)
        print("your account has been credited : ",amt)
        print(f'total balance is {self.get_balance()}')
    def debit(self,amt):
        self.balance-=amt
        print("your account number is : ",self.account_no)
        print("your account has been debited : ",amt)
        print(f'total balance is {self.balance}')
    def get_balance(self):
        return self.balance
        
        
person1=Account(1000,1234567) 
person1.debit(500) 
person1.credit(1500) 
person1.credit(5500) 
person1.debit(5000) 
person1.get_balance()


   
```

```python
class BankAccount:
    def __init__(self, owner, balance,__pin):
        self.owner = owner
        self.balance = balance
        self.__pin = __pin

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}, New Balance: {self.balance}")

    def withdraw(self, amount):
        try:
            pin = int(input("Enter your PIN: "))
            if pin == self.__pin:
                if amount <= self.balance:
                    self.balance -= amount
                    print(f"Withdrawn {amount}, New Balance: {self.balance}")
                else:
                    print("Insufficient funds")
            else:
                print("Incorrect PIN")
        except:
            print(f"enter the correct pin")     
class SavingsAccount(BankAccount):
    def add_interest(self, rate):
        interest = self.balance * rate
        self.deposit(interest)
        print(f"Interest Added: {interest}")

# Example usage
john = SavingsAccount("John Doe", 1000, 1234)
john.withdraw(500)
```
