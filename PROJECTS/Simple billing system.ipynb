```python
#### Simple billing system
print("-------------------bill---------------------")
size=int(input("Enter the product list size:"))
product=[]
price=[]
for i in range(size):
    i=input("enter product name:-")
    product.append(i)

for i in range(size):
    i=int(input("Enter product price:"))
    price.append(i)
    

total=0
for i in range(size):
    total=total+price[i]
    print((i+1),product[i],"=",price[i],"/-")
print("-----------------")
print("Total amount:-",total,"/-")
```
